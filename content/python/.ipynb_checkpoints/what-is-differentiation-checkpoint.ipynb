Calculus is one of the essential pillars of a good quantitative skillset (alongside linear algebra, statistics and geometry). That being said, many people who learn it don't really get to grasp the power and simplicity of the discipline. 

At it's core, Calculus is the study of limits; it's a way of thinking about very abstract and challenging concepts and making them something we can get our hands--and heads--around. As such, Calculus really has two main tricks: differentiation and integration and the two are quite related to one another. 

In this post, we're going to talk about differentiation and give an introduction to what it is and how we can use it. In the process, we're going to use the Python library `sympy` which will allow us to work with symbolic math thereby making it easy to explore differentiation in greater depth.

We'll start by importing sympy and numpy

and we'll use sympy to create a symbol "x" which, as it sounds, allows us to work with it as a symbol meaning we dont have to declare any particular value for it

```python
x = sm.symbols('x')
```

We can start simply by plotting some function of x which should be familiar to most: 

$$y = x^2$$ 

```python
y = x**2
```

Now to understand differentiation, let's imagine a pseudo real world scenario: Let's say that we have a car and the y-axis represents the distance our car has travelled (the odometer reading) and the x-axis represents time so we have a distance traveled for each point in time. 

We can appreciate that the slope of our line represens a change in distance traveled (y) over some time (x) and, this represents velocity. We can also appreciate that our velocity seems to be increasing as time goes on. We feel like this means the car is accellerating but what if we wanted to know *by exactly how much* it was accellerating? Is there a mathematical way to get this sort of information from the function? Well there is and it's differentiation that allows us to do this. Let's explore that:

Many students have taken calculus at some point in high school or college and learned a definition of differentiation that's somewhat paradoxical: "the instantaneous rate of change" of a function. If you pause for a moment to think on that, though, it's rather troubling. How can there be such a thing as an instantaneous rate of change? Doesn't all change happen *over time*? I said I wanted to know how much the velocity was changing at a given point in time, is my question nonsense? This line of questioning is really why Calculus exists. It allows us to get at such a concept without getting hung up on its deep logical and philosophical issues. The truth is that there actually is no answer to the question "how much is the speed of the car changing at a given instant" but we can *approximate* such an answer through differentiation and this whole theme of approximating things is really the heart and soul of calculus.

So let's start simple: what if, instead of asking about an *instant* in time, we just asked about just a *small* interval in time? Let's first write a function that will take in two points and return our line function

```python
def get_line_between_x_points(start, finish, y):
    x = sm.symbols('x')
    y_func = sm.lambdify(x, y, 'numpy')
    
    print(y_func(start))
    
    x1, y1 = start, y_func(start)
    x2, y2 = finish, y_func(finish)
    
    slope = (y2 - y1) / (x2 - x1)
    
    intercept = y1 - (slope*x1)
    
    return sm.sympify(slope*x + intercept)
    
```

Now we'll use that function to add the line between two given points to our plot. Let's looks at points x = 1 and x = 4 for example:

```python
line1 = get_line_between_x_points(1, 5, y)

p2 = sm.plot(y, line1, (x,0,5), show=False)
p2[1].line_color = 'red'
p2.show()
```

we notice here that we get a sort of crude understanding of the rate of change of our blue line but we lose some granualrity. For example, we get that the line has an upward slope (so we're moving) but, if all we had was the red line, we would probably think that we weren't actually accellerating. But what happens if we shrink that window? Let's say we're particularly interested in what happens at x=2, so let's shrink our window a bit around that number:

```python
line1 = get_line_between_x_points(3, 4, y)

p2 = sm.plot(y, line1, (x,0,5), show=False)
p2[1].line_color = 'red'
p2.show()
```

this is better for sure because the red line now seems to hug closer to the blue line at our point of interest. In other words, for our point x=2, the slope of the red line looks more like the slope of the blue line.

Let's shrink the range of x again:

```python
line1 = get_line_between_x_points(3.49, 3.51, y)

p2 = sm.plot(y, line1, (x,0,5), show=False)
p2[1].line_color = 'red'
p2.show()
```

... even better! Our range of x is getting pretty small (between 1.99 and 2.01) and, in reality, we could keep going, making it smaller and smaller still. 

If we follow that logic, what we're really asking for is the slope of a line between to points that are closer and closer together. That it to say we want the *change in y* for a given *change in x* as x approaches 0. As we said earlier, there is no answer to that question when the change in x actually *is* 0 but we want to estimate what it might be, what the function would *behave like* as the change in x approaches 0.

In more mathematical terms, we would call this process *taking the derivative* of our function y and we would refer to the derivative as $y \prime$ and, what's more, we think of this derivative as being *with respect* to some variable, in this case x. 

So what we're doing here is definining the derivative of some function $y = f(x)$ as $y \prime = lim_{x \to 0} ; {dy} / {dx}$ or "the change in y for a given change in x as that change in x (the space between points) gets smaller and smaller and approaches 0.

And interestingly, the derivative of a function is a function itself. In this case, for the function $y = x^2$, the derivative is $y \prime = 2x$

```python
y_prime = y.diff(x)
p3 = sm.plot(y, line1, y_prime, (x,0,5), show=False)
p3[1].line_color = 'red'
p3[2].line_color = 'green'
p3.show()
```

Now if we plot our function in blue, a line through two points very close to 2 in red (as we did above) and the proper derivative in green, we see something interesting: that the green line shows us the slope of our blue line at each given value of x. For x = 3.5, the green line has a value of y = 7 which tells us that the value of the blue line at x = 3.5 is 7, which is what the red line is approaching as its range of x gets smaller and smaller.

As one final illustration, we can take a window over x and plot a line at each position for that window. We get an instresting output (all the red lines) that really illustrates the change in 

```python
def make_colormap(num_colors):
    cm = pyl.get_cmap('Reds')

    colors = [ cm(1.*i/num_colors) for i in range(num_colors) ]
    return colors
```

```python
x_max = 5
x_min = 0

lines = []

for i in range(6):
    
    x1 = x_min + i
    x2 = x1 + 1
    
    lines.append(get_line_between_x_points(x1, x2, y))
```

```python
p4 = sm.plot(y, (x,0,5), show=False)

for line,color in zip(lines,make_colormap(len(lines))):
    thing = sm.plot(line, (x,0,5), show=False, line_color=color)
    p4.append(thing[0])

p4.append(sm.plot(y, (x,0,5), show=False)[0])
p4.append(sm.plot(y_prime, (x,0,5), show=False, line_color='green')[0])
p4.show()
```
